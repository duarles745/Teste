import matplotlib.pyplot as plt é um comando do python para importar a biblioteca matplotlib que sera utilizada para gerar a interface gráfica do gráfico

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(8,5))
plt.title("Função Constante")

x = np.linspace(-10, 10, 10000)
y = 0.8*np.ones(x.shape)

plt.plot(x, y)
plt.show()
```
